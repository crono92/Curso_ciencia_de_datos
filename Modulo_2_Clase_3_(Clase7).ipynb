### 2- Calculo de edad

```python
df_clientes["fecha_nacimiento"]
```

```python
from datetime import datetime, date

# Convertir la columna 'Fecha_Nacimiento' al tipo de datos datetime
df_clientes["fecha_nacimiento"] = pd.to_datetime(df_clientes["fecha_nacimiento"])
df_clientes["fecha_nacimiento"]
```

```python
# Calcular la edad del usuario
import numpy as np

ahora = date.today()
df_clientes['edad'] = (ahora.year - df_clientes["fecha_nacimiento"].dt.year)

# Mostrar el DataFrame con la nueva columna 'edad'
print(df_clientes)
```

```python
print("La edad promedio es de ", df_clientes['edad'].mean()," años")
print("El cliente mayor tiene ", df_clientes['edad'].max()," años")
print("El cliente menor tiene ", df_clientes['edad'].min()," años")
```

### Fusionar clientes con boletas usando el rut

```python
print("Ruts en boletas:",len(df_boletas['rut_cliente'].unique()))
print("Cantidad de ruts en listado clientes",len(df_clientes["rut"].unique()))
```
